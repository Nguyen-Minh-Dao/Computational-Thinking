### W4A6

```python
import math
a,b = map(int, input().split())
prime = [True] * (b + 1)
prime[0] = prime[1] = False
for i in range(2, math.isqrt(b) + 1):
    if prime[i]:
        for j in range(i**2, b + 1, i):
            prime[j] = False
sum = 0
for i in range(a, b + 1):
    if prime[i]:
        sum += i
print(sum)
```

### W4A8

```python
def reverse(a):
    b = 0
    while a > 0:
        b = b * 10 + a % 10
        a //= 10
    return b
n = int(input())
step = 0
while n != reverse(n):
    n += reverse(n)
    step += 1
print(step, n)
```

### W4A13

```python
import math
def dsum(n):
    if n <= 1:
        return 0
    ans = 1
    for i in range(2, math.isqrt(n) + 1):
        if n % i == 0:
            ans += i
            if n /i != i:
                ans += n // i
    return ans
a,b = map(int, input().split())
print(dsum(a) == b and dsum(b) == a)
```
