### 16

```python
n = float(input())
if n == int(n):
    print(n, n, n)
elif n > 0:
    if n - int(n) >= 0.5:
        print(int(n) + 1, int(n), int(n) + 1)
    else:
        print(int(n) + 1, int(n), int(n))
else:
    if int(n) - n >= 0.5:
        print(int(n), int(n) - 1, int(n) - 1)
    else:
        print(int(n), int(n) - 1, int(n))
```
