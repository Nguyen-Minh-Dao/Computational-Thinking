### W6A16
Nhậ̣p
vào mảng n hàng, m cột số nguyên. In lại mảng ra màn hình như ma
trận hai chiều, định dạng mỗi phần tử 4 vị trí căn phải

```python
n, m = map(int,input().split())
ans = []
for i in range(n):
    row = []
    for j in range(m):
        x = int(input())
        row.append(x)
    ans.append(row)
for i in range(n):
    for j in range(m):
        print(f"{(ans[i][j]):>4}", end = " ")
    print()
```

### W6A17
Nhập vào mảng vuông 2 chiều n hàng, n cột, in ra đường chéo chính và
đường chéo phụ của mảng

```python
n = int(input())
matrix = []
for i in range(n):
    row = []
    for j in range(n):
        x = int(input())
        row.append(x)
    matrix.append(row)
for i in range(n):
    print(matrix[i][i], end = " ")
print()
for i in range(n):
    print(matrix[i][n-i-1], end = " ")
```

### W6A18
Nhậ̣p
vào mảng n hàng, m cột số nguyên và số nguyên k (n, m, k nhập trên
cùng một dòng, cách nhau dấu cách). Tính và in ra tổng các phần tử trên cột
k (hàng k)

```python
n, m, k = map(int, input().split())
matrix = []
col_sum, row_sum = 0, 0
for i in range(n):
    row = []
    for j in range(m):
        x = int(input())
        row.append(x)
    matrix.append(row)
for i in range(n):
    row_sum += matrix[i][k - 1]
for i in range(m):
    col_sum += matrix[k - 1][i]
print(row_sum, col_sum)
```

### W6A19
Viết hàm đưa vào 1 list số nguyên, tìm và trả về vị trí có giá trị lớn nhất
trong list

### W6A20
Viết hàm đưa vào một list số nguyên và một số nguyên dương k. Hãy tìm và
trả về vị trí của phần tử đầu tiên có giá trị k trong list, nếu không có thì trả về
-1

```python
nums = list(map(int, input().split()))
k = int(input())
if k not in nums: print(-1)
else:
    for i, n in enumerate(nums):
        if n == k:
            print(i)
            break
```

### W6A21
Nhập một dòng gồm các chuỗi (tách bởi dấu cách). Tạo dictionary ánh xạ
chuỗi -> tuple các vị trí (0-based) mà chuỗi đó xuất hiện

```python
s = input().split()
ans = {}
for i, c in enumerate(s):
    if c in ans:
        ans[c].append(i)
    else:
        ans[c] = [i]
for i in ans:
    ans[i] = tuple(ans[i])
print(ans)
```

### W6A22
Viết hàm đưa vào 1 list số nguyên, tìm và trả về vị trí có giá trị lớn nhất
trong list

```python
def solve(arr:list) -> int:
    for i, n in enumerate(arr):
        if n == max(arr):
            return i
nums = list(map(int,input().split()))
solve(nums)
```

### W6A23
Viết hàm xóa đi các số <x có trong mảng

```python
def solve(arr:list, n:int) -> list:
    ans = []
    for i in arr:
        if i >= n:
            ans.append(i)
    return ans
nums = list(map(int,input().split()))
x = int(input())
solve(nums, x)
```

### W6A27

### W6A28
Cho một dãy các số nguyên, kiểm tra xem nó là dãy tăng thực sự hay giảm
thực sự, hay vô danh (không tăng, không giảm)

```python
nums = list(map(int,input().split()))
if nums == sorted(nums):
    print("Tăng thực sự")
elif nums == sorted(nums, reverse=True):
    print("Giảm thực sự")
else:
    print("Vô danh")
```
