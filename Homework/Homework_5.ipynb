### W5A16

```python
def AmstrongCheck(x:int)->bool:
    if x < 0: return False
    s = str(x)
    length = len(s)
    return x == sum(int(digit)**length for digit in s)
```

### W5A17

```python
def SquareCheck(x:int)->bool:
    if x < 0: return False
    i = int(x**0.5)
    return i*i == x
```

### W5A18

```python
def PalindromeCheck(x:int)->bool:
    if x < 0: return False
    og_x = x
    palin_x = 0
    while og_x != 0:
        palin_x = palin_x * 10 + og_x % 10
        og_x //= 10
    return palin_x == x
```

### W5A19

```python
def MagicCheck(x:int)->bool:
    x = abs(x)
    while x >= 10:
        x = sum(int(digit) for digit in str(x))
    return x == 1
```

### W5A20

```python
def detect_number_type(s:str)->str:
    try:
        int(s)
        return "int"
    except ValueError:
        try:
            float(s)
            return "float"
        except ValueError:
            return "neither"
```

### W5A23

```python
def quadratic_roots_count(a:float,b:float,c:float)->int:
    if a == 0 and b == 0:
        return 0 if c != 0 else float('inf')
    discriminant = b**2 - 4*a*c
    if discriminant > 0:
        return 2
    elif discriminant == 0:
        return 1
    else:
        return 0
```

### W5A25

```python
import math
def point_in_quadrilateral(quad:list,p:tuple):
    x, y = p
    inside = False
    n = len(quad)
    for i in range(n):
        p1x, p1y = quad[i]
        p2x, p2y = quad[(i + 1) % n]
        crossproduct = (y - p1y) * (p2x - p1x) - (x - p1x) * (p2y - p1y)
        if math.isclose(crossproduct, 0, abs_tol=1e-9):
            dotproduct = (x - p1x) * (p2x - p1x) + (y - p1y) * (p2y - p1y)
            squared_length = (p2x - p1x)**2 + (p2y - p1y)**2
            if 0 <= dotproduct <= squared_length:
                return True
        if (p1y > y) != (p2y > y):
            x_intersect = (p2x - p1x) * (y - p1y) / (p2y - p1y) + p1x if p2y != p1y else p1x
            if x < x_intersect:
                inside = not inside
    return inside
```

```python
def nonzero_digits_cnt(nums:list)->int:
    cnt = 0
    for n in nums:
        for digit in str(n):
            if digit != '0':
                cnt += 1
    return cnt
```

### W5A28

```python
import math
def Angle(x1,y1,x2,y2,x3,y3)->float:
    a = math.sqrt((x2 - x3)**2 + (y2 - y3)**2)
    b = math.sqrt((x1 - x3)**2 + (y1 - y3)**2)
    c = math.sqrt((x1 - x2)**2 + (y1 - y2)**2)
    if a == 0 or b == 0: return 0.0
    cos_angle = (a**2 + b**2 - c**2) / (2 * a * b)
    cos_angle = max(-1, min(1, cos_angle))
    return math.degrees(math.acos(cos_angle))
```
