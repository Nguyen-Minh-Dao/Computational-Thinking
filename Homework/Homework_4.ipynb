### W4A22

```python
n = int(input())
even, odd = 0
while n > 0:
    if n % 10 % 2 == 0:
        even += 1
    else:
        odd += 1
    n //= 10
print(even, odd)
```

### W4A27

```python
s = input()
length = sum(1 for _ in s)
word = 0
for i in range(length - 1):
    if s[i] != ' ' and s[i + 1] == " ":
        word += 1
print(word + 1)
```

### W4A28

```python
s = input()
length = sum(1 for _ in s)
word = ""
for i in range(length - 1):
    if s[i] == " ":
        break
    else:
        word += s[i]
print(word)
```
