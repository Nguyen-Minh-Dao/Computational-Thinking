### 7

```python
import math
n = int(input())
if n == 1:
    print(1)
else:
    cnt = 2
    for i in range(2, math.isqrt(n) + 1):
        if n % i == 0:
            cnt += 1
            if (n / i) != i:
                cnt += 1
    print(cnt)
```

### 14

```python
import math
ans_max = -math.inf
ans_min = math.inf
while True:
    n = int(input())
    if ans_max < n:
        ans_max = n
    if ans_min > n:
        ans_min = n
    if n == -1:
        break
print(ans_max, ans_min)
```

```python
n = int(input())
even = 0
odd = 0
while n != 0:
    i = n % 10
    if i % 2 == 0:
        even += 1
    else:
        odd += 1
    n //= 10
print(f"{even} even digits, {odd} odd digits")
```

### 20

```python
def PrimeCheck(a):
    if a == 1:
        return False
    if a == 2:
        return True
    for i in range(2, int(a**0.5) + 1):
        if a % i == 0:
            return False
    return True
n = int(input())
cnt = 0
while n != 0:
    i = n % 10
    if PrimeCheck(i):
        cnt += 1
    n //= 10
print(cnt)
```
