### W6A4

```python
k_v = input().split()
res = {}
for i in k_v:
    k, v = i.split(':')
    if k not in res:
        res[k] = []
    res[k].append(v)
print(res)
```

### W6A5

```python
nums = list(map(int, input().split()))
res = {'positives':0, 'negatives':0, 'zeros':0}
for i in nums:
    if i > 0:
        res['positives'] += 1
    elif i < 0:
        res['negatives'] += 1
    else:
        res['zeros'] += 1
print(res)
```

### W6A10

```python
nums = map(int, input().split())
k = int(input())
seen = set()
ans = set()
for i in nums:
    if k - i in seen:
        ans.add((k - i, i))
    seen.add(i)
ans = sorted(list(ans))
print(ans)
```

### W6A11

```python
nums = tuple(map(int, input().split()))
even_nums = tuple(x for x in nums if x % 2 == 0)
odd_nums = tuple(x for x in nums if x % 2 != 0)
print(even_nums)
print(odd_nums)
```

### W6A12

```python
nums = list(map(int, input().split()))
cnt = {}
for i in nums: cnt[i] = cnt.get(i, 0) + 1
max_cnt = max(cnt.values())
ans = [i for i, cnt in cnt.items() if cnt == max_cnt]
print(min(ans))
```

### W6A15

```python
s = input().split()
k = int(input())
ans = {}
for key, value in zip(s[0::2], s[1::2]):
    if int(value) > k:
        ans[key] = int(value)
print(ans)
```
