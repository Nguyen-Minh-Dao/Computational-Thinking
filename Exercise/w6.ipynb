### W6A2

```python
nums = list(map(int, input().split()))
res = []
curr = 0
for i in nums:
    curr += i
    res.append(curr)
print(res)
```

### W6A3

### W6A4

```python
k_v = input().split()
res = {}
for i in k_v:
    k, v = i.split(':')
    if k not in res:
        res[k] = []
    res[k].append(v)
print(res)
```

### W6A5

```python
nums = list(map(int, input().split()))
res = {'positives':0, 'negatives':0, 'zeros':0}
for i in nums:
    if i > 0:
        res['positives'] += 1
    elif i < 0:
        res['negatives'] += 1
    else:
        res['zeros'] += 1
print(res)
```
