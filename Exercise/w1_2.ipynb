```python
# Duyệt qua một danh sách các số và kiểm tra xem từng số trong danh sách là số chẵn hay số lẻ
numbers = [1, 2, 3, 4]
for x in numbers:
    if x % 2 == 0:
        print(x, "là số chẵn")
    else:
        print(x, "là số lẻ")
```

### W1A7

```python
num = input("Nhập số: ")
if num % 2 = 0:
    print("Số chẵn")
else
    print("Số lẻ")
# Lỗi kiểu dữ liệu: input() trả về chuỗi văn bản (string). Không thể dùng toán tử % trực tiếp với chuỗi, phải ép kiểu về số nguyên (int).
# Lỗi toán tử so sánh: Ở dòng if num % 2 = 0, toán tử = là phép gán. Để so sánh bằng, phải sử dụng ==
# Lỗi cú pháp: Thiếu dấu hai chấm : ở cuối từ khóa else
num = int(input("Nhập số: "))  
if num % 2 == 0:               
    print("Số chẵn")
else:                          
    print("Số lẻ")
```
