### W5A3

```python
def PrimeCheck(n:int):
    if n <= 1:
        return False
    if n == 2:
        return True
    for i in range(2, int(n**0.5) + 1):
        if n % i == 0:
            return False
    return True
```

### W5A4

```python
def PerfectCheck(n:int):
    if n <= 0:
        return False
    sum_divisors = 0
    for i in range(1, n):
        if n % i == 0:
            sum_divisors += i
    return sum_divisors == n
```

### W5A5

```python
from typing import List
def First_k(lst: List[int], k: int):
    try:
        return lst.index(k)
    except ValueError:
        return -1
```

### W5A6

```python
def Factorial(n):
    result = 1
    for i in range(2, n + 1):
        result *= i
    return result
```

### W5A7

```python
def Calc(num1: float, num2: float, operator: str):
    if operator == '+':
        result = num1 + num2
    elif operator == '-':
        result = num1 - num2
    elif operator == '*':
        result = num1 * num2
    elif operator == '/':
        if num2 == 0:
            raise ZeroDivisionError()
        result = num1 / num2
    return f"{result:.2f}"
```

### W5A8

```python
def Hamming(x: int, y: int):
    return bin(x ^ y).count('1')
```

### W5A9

```python
def DigitSum(n: int):
    return sum(int(digit) for digit in str(n))
```

### W5A10

```python
def Isomorphic(a: str, b: str):
    if len(a) != len(b):
        return False
    return len(set(a)) == len(set(b)) == len(set(zip(a, b)))
```

### W5A12

```python
from typing import Tuple
def divide_with_remainder(dividend: int, divisor: int) -> Tuple[int, int]:
    if divisor == 0:
        raise ValueError()
    return divmod(dividend, divisor)
```

### W5A15

```python
def is_right_triangle(a: float, b: float, c: float):
    if a <= 0 or b <= 0 or c <= 0:
        return False 
    sides = sorted([a, b, c])
    import math
    return math.isclose(sides[0]**2 + sides[1]**2, sides[2]**2)
```
